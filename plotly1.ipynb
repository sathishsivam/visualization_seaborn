```python
import plotly.express as px
```

```python
df = px.data.iris()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species")
```

```python
fig
```

```python
import seaborn as sns
```

```python
sns.relplot(x="sepal_width",y="sepal_length",data=df,hue="species")
```
