```python
import seaborn as sns
```

```python
data=sns.load_dataset("tips")
```

```python
data
```

single numeric variable shows its univariate distribution with points randomly “jittered” on the other axis:

```python
sns.stripplot(x="total_bill",data=data)
```

Assigning a second variable splits the strips of points to compare categorical levels of that variable:

```python
sns.stripplot(x="total_bill",y="day",data=data)
```

```python
sns.stripplot(x="day",y="total_bill",data=data)
```

```python
sns.stripplot(x="day",y="total_bill",data=data,hue="size",jitter=True)
```

```python
sns.stripplot(data=data)
```
